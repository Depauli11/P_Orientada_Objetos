1. Crie uma classe chamada Retângulo. Um retângulo possui uma base e uma altura.
Crie os métodos necessários para que o usuário possa obter informações sobre a
base, a altura e a área do retângulo.


```python
class Retangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
                
    def area_retangulo(self):
        self.area = (self.base * self.altura)/2
        return self.area


def main():
    a = Retangulo(2,6)
    print(a.altura)
    print(a.base)
    print(a.area_retangulo())
    

if __name__ == '__main__':
    main()
```

2. Crie uma classe chamada Pessoa. Uma pessoa possui nome, idade, altura,
quantidade de irmãos e endereço. 

Na classe Pessoa:

• crie um método chamado imprime_info, que imprime na tela as informações
da pessoa, de maneira organizada.

• crie um método chamado is_filho_unico, que retorna verdadeiro caso a
pessoa seja filha única e falso caso contrário.

No método main() faça o que se pede:

• crie 3 pessoas informando todos os dados.

• imprima as informações de todas as pessoas, de forma legível e organizada.

• imprima na tela a frase “Filho(a) único(a)” para as pessoas que forem filhas
únicas, e a frase “Não é filho(a) único(a)” para as pessoas que não forem filhas
únicas.


```python
class Pessoa:
    def __init__(self, nome, idade, altura, qty_irmaos, endereco):
        self.nome = nome
        self.idade = idade
        self.altura = altura
        self.qty_irmaos = qty_irmaos
        self.endereco = endereco

    def imprime_info(self):
        print(f'Informações de {self.nome}')
        print(f'Idade: {self.idade}')
        print(f'Altura: {self.altura}')
        print(f'Quantidade de irmãos: {self.qty_irmaos}')
        print(f'Endereço: {self.endereco}')


    def is_filho_unico(self):
        if self.qty_irmaos > 0:
            return print('Não é filho(a) único(a)\n')
        else:
            return print('Filho(a) único(a)\n')


def main():
    pessoa01 = Pessoa('Marcos',34,1.76,0,'rua dos bolos, 0' )
    pessoa01.imprime_info()
    pessoa01.is_filho_unico()

    pessoa02 = Pessoa('Leticia', 65, 3.56,98,'beco sem saída')
    pessoa02.imprime_info()
    pessoa02.is_filho_unico()

    pessoa03 = Pessoa('Carlos', 45, 1.65,4,'Avenida Unisinos')
    pessoa03.imprime_info()
    pessoa03.is_filho_unico()



if __name__ == '__main__':
    main()
```

3. Crie uma classe para calcular o imposto de renda simplificado chamada
IRSimplificado. A classe IRSimplificado possui o nome do contribuinte, o ano de
nascimento, a profissão, a escolaridade, a renda mensal e o número de
dependentes do contribuinte. 

Crie os seguintes métodos:

• método que calcula e retorna a idade do contribuinte

• método que calcula e retorna a renda anual do contribuinte

• método que calcula e retorna a renda per capita mensal

• método que retorna a alíquota de IR máxima

• método que retorna o valor da dedução do IR segundo a alíquota de IR máxima

• método que calcula e retorna o valor do imposto de renda mensal

• método que calcula e retorna o valor do imposto de renda anual

Respeitar as faixas para as alíquotas segundo a tabela abaixo.

Escreva um programa que instancie um objeto de forma dinâmica e solicite que o
usuário informe todos os atributos. Em seguida, faça o cálculo do imposto de renda
e imprima todos os dados.

```python
class IRSimplificado:
    def __init__(self, contribuinte, ano_nascimento, profissao, escolaridade, renda, qty_dependentes):
        self.contribuinte = contribuinte
        self.ano_nascimento = ano_nascimento
        self.profissao = profissao
        self.escolaridade = escolaridade
        self.renda = renda
        self.qty_dependentes = qty_dependentes

    def idade(self):
        self.idade = 2024-self.ano_nascimento
        return self.idade
    
    def renda_anual(self):
        self.renda_anual = self.renda*12
        return self.renda_anual
    
    def renda_per_capita(self):
        self.renda_per_capita = self.renda/self.qty_dependentes
        return self.renda_per_capita
    
    def aliquota_maxima(self):
        if self.renda <= 2259.20:
            return 0.0

        if self.renda > 2259.20 and self.renda <= 2826.65 :
            return 0.075

        if self.renda > 2826.65 and self.renda <=  3751.05 :
            return 0.15

        if self.renda > 3751.05 and self.renda <= 4664.68 :
            return 0.225

        if self.renda > 4664.68:
            return 0.275



    def deducao_aliquota_maxima(self):
        if self.aliquota_maxima() == 0:
            return 0.0

        if self.aliquota_maxima() == 0.075:
            return 169.44

        if self.aliquota_maxima() == 0.15:
            return 381.44

        if self.aliquota_maxima() == 0.225:
            return 662.77

        if self.aliquota_maxima() == 0.275:
            return 896.00

    
    def imposto_mensal(self):
        return self.renda*self.aliquota_maxima() - self.deducao_aliquota_maxima()

        
    def imposto_anual(self):
        self.imposto_anual = self.imposto_mensal()*12
        return self.imposto_anual
    

    def imprime_all(self):
        print(f'Resumo Imposto de Renda\n')
        print(f'Nome: {self.contribuinte}\n')
        print(f'Ano de Nascimento: {self.ano_nascimento}\n')
        print(f'Profissão: {self.profissao}\n')
        print(f'Escolaridade: {self.escolaridade}\n')
        print(f'Renda Mensal: {self.renda}\n')
        print(f'Número de dependentes: {self.qty_dependentes}\n')
        print(f'Alíquota máxima: {self.aliquota_maxima()}\n')
        print(f'Dedução da alíquota máxima: {self.deducao_aliquota_maxima()}\n')
        print(f'Imposto mensal: {self.imposto_mensal():.2f}\n')
        print(f'Imposto anual: {self.imposto_anual():.2f}\n')



def main():
    ir_pessoa01 = IRSimplificado(input(f"Informe o nome do contribuinte: "),
                                 int(input(f"Informe o ano de nascimento do contribuinte: ")),
                                 input(f"Informe a profissão do contribuinte: "),
                                 input(f"Informe a escolaridade do contribuinte: "),
                                 float(input(f"Informe a renda mensal do contribuinte: ")),
                                 int(input(f"Informe o número de dependentes do contribuinte: ")))
    
    ir_pessoa01.imprime_all()
                                 



if __name__ == '__main__':
    main()



    
```
