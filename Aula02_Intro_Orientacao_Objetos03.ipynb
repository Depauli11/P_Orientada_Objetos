1. Classe Calculadora:

◦ Uma calculadora possui uma memória e uma cor.

◦ Quando uma calculadora é criada, a memória deve ser inicializada com 0 e a cor
deve ser recebida por parâmetro (construtor).

◦ Crie os métodos de acesso para os atributos da classe (get e set).

◦ Crie os métodos: soma, subtrai, multiplica e divide. Todos recebem dois
valores (float) por parâmetro e retornam o valor da operação realizada.

◦ Crie os métodos eleva_ao_quadrado e eleva_ao_cubo. Ambos recebem
apenas um valor (int) e retornam o valor da operação realizada.

◦ Crie um método imprime_info, que não recebe parâmetros e simplesmente
imprime as informações da calculadora de maneira legível e organizada.


```python
class Calculadora:
    def __init__(self, cor):
        self.__memoria = 0
        self.__cor = cor

    @property
    def memoria(self):
        return self.__memoria
    
    @property
    def cor(self):
        return self.__cor
    
    @memoria.setter
    def memoria(self, memoria):
        self.__memoria = memoria

    @cor.setter
    def cor(self, cor):
        self.__cor = cor

    def soma(self, valor01:float, valor02:float):
        return valor01 + valor02
    
    def subtrai(self, valor01:float, valor02:float):
        return valor01 - valor02
    
    def multiplica(self, valor01:float, valor02:float):
        return valor01 * valor02
    
    def divide(self, valor01:float, valor02:float):
        return valor01 / valor02
    
    def eleva_ao_quadrado(self, valor01:int ):
        return valor01**2
    
    def eleva_ao_cubo(self, valor01:int ):
        return valor01**3

    def imprime_info(self):
        print(f'Informações da Calculadora')
        print(f'Memória: {self.memoria}')
        print(f'Cor: {self.cor}\n')

```

2. Classe FuncionarioCaixa:

◦ Possui um nome, um endereço e uma calculadora.

◦ Crie um construtor que recebe os parâmetros para inicializar todos os atributos.

◦ Crie os métodos de acesso dos atributos desta classe (get e set).

◦ Crie os métodos soma, subtrai, multiplica, divide, eleva_ao_quadrado e eleva_ao_cubo. Cada método destes deve chamar o
método correspondente da calculadora, retornando o valor obtido na operação.

◦ Crie um método chamado imprime_info, que não recebe parâmetros de entrada e imprime na tela as informações da classe, inclusive da calculadora (sem as operações, apenas os atributos).



```python
class FuncionarioCaixa:
    def __init__(self, nome, endereco, calculadora:Calculadora):
        self.__nome = nome
        self.__endereco = endereco
        self.__calculadora = calculadora

    @property
    def nome(self):
        return self.__nome
    
    @property
    def endereco(self):
        return self.__endereco
    
    @property
    def calculadora(self):
        return self.__calculadora
    
    @nome.setter
    def nome(self, nome):
        self.__nome = nome 
    
    @endereco.setter
    def enderco(self, endereco):
        self.enderco = endereco

    @calculadora.setter
    def calculadora(self, calculadora):
        self.calculadora = calculadora

    def soma(self, valor01, valor02 ):
        return self.__calculadora.soma(valor01, valor02)
    
    def subtrai(self, valor01, valor02):
        return self.__calculadora.subtrai(valor01, valor02)
    
    def multiplica(self, valor01, valor02):
        return self.__calculadora.multiplica(valor01, valor02)
    
    def divide(self, valor01, valor02):
        return self.__calculadora.divide(valor01, valor02)
    
    def imprime_info(self):
        print(f'Informações do Funcionário')
        print(f'Nome: {self.nome}')
        print(f'Endereço: {self.enderco}')
        self.calculadora.imprime_info()
```

Classe Empresa:

◦ Uma empresa tem um nome e dois funcionários do caixa.

◦ Crie um construtor que recebe todos os parâmetros para inicializar os atributos.

◦ Crie os métodos de acesso dos atributos desta classe (get e set).

◦ Crie um método imprime_info, que imprime as informações da classe.

```python
class Empresa:
    def __init__(self, nome, funcionario01:FuncionarioCaixa, funcionario02:FuncionarioCaixa):
        self.__nome = nome
        self.__funcionario01 = funcionario01
        self.__funcionario02 = funcionario02

    @property
    def nome(self):
        return self.__nome
    
    @property
    def funcionario01(self):
        return self.__funcionario01
    
    @property
    def funcionario02(self):
        return self.__funcionario02
    
    @nome.setter
    def nome(self, nome):
        self.__nome = nome

    @funcionario01.setter
    def funcionario01(self, funcionario01):
        self.__funcionario01 = funcionario01

    @funcionario02.setter
    def funcionario02(self, funcionario02):
        self.__funcionario02 = funcionario02

    def imprime_info(self):
        print(f'Nome da empresa: {self.nome}\n')
        self.funcionario01.imprime_info()
        self.funcionario02.imprime_info()
```

devem ser feitas através da calculadora do objeto funcionário criado.

funcionário que deve ser criado. Tudo que for necessário para criar esta empresa
(que já não tenha sido criado) deve ser solicitado para o usuário.

• Imprima as informações desta empresa.
